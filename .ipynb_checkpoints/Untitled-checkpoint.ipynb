```python
import numpy as np
import pandas as pd
import scipy
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import figure
%matplotlib inline

df = pd.read_csv("./Bitcoin Price (2014-2023)_exported.csv")
df.head()
```

**Добавление нового атрибута**

Добавим атрибут процентное изменение (Percentage Change):
Это колонка, показывающая процентное изменение цены закрытия по сравнению с предыдущим днем.

```python
import pandas as pd

df = pd.read_csv("./Bitcoin Price (2014-2023)_exported.csv")
df['Close'] = pd.to_numeric(df['Close'], errors='coerce')
df['Percentage Change'] = df['Close'].pct_change() * 100
print(df.columns)
df['Percentage Change'] = df['Percentage Change'].fillna(0)
```

```python
df.head()
```

**"Причёсывание датасета"**

Обработка пропущенных значений

```python
df.isnull().sum()
```

Определяем количество выбросов

```python
columns = ['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume', 'Percentage Change']

def plot_histograms_and_detect_outliers(df, columns):
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1

        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]

        print(f'Для {col}:')
        print(f'Нижняя граница выбросов: {lower_bound}')
        print(f'Верхняя граница выбросов: {upper_bound}')
        print(f'Количество выбросов: {outliers.shape[0]}')

plot_histograms_and_detect_outliers(df, columns)
```

```python
import pandas as pd
import numpy as np

numeric_cols = df.select_dtypes(include=[np.number]).columns

df_filtered = df.copy()

for col in numeric_cols:
    Q1 = df_filtered[col].quantile(0.25)
    Q3 = df_filtered[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df_filtered = df_filtered[(df_filtered[col] >= lower_bound) & (df_filtered[col] <= upper_bound)]
df_filtered.info()
```

**Построение графиков**

Кластеры которые можно выделить:

y=x (где цены открытия близки к ценам закрытия), это может указывать на стабильные рыночные условия, где цена актива не сильно колеблется в течение торгового дня. Такое поведение можно наблюдать между всеми параметрами кроме обьема и процентного изменения цены закрытия по сравнению с предыдущим днем.Это может указывать на  рыночные условия, где цена актива не сильно колеблется. 

Точки расположены более хаотично у отношений между объемами и процентного изменения цены. 
На графике рассеяния, отображающем процентное изменение цены закрытия (Percentage Change), наблюдается такое распределение, где много точек сосредоточено в начале по оси X (что может указывать на небольшие или отрицательные изменения), а по оси Y точки равномерно распределены, это может указывать на наличие долгосрочного тренда, где в разные дни наблюдаются как положительные, так и отрицательные изменения. Это может означать, что актив периодически демонстрирует значительные колебания, но в целом остается в пределах определенного диапазона.

на графике рассеяния объема наблюдается, что в начале точки расположены вдоль линии y=x, а далее распределение становится более хаотичным, с уменьшением количества точек по мере удаления от нуля. Распределение точек вдоль линии y=x в начале может указывать на сильную корреляцию между объемом торгов и процентным изменением цены на начальных этапах , когда цена меняется незначительно, объем остается на стабильном уровне. Хаотичное распределение точек за пределами начальной области может указывать на более сложные рыночные условия, где влияние различных факторов приводит к непредсказуемым изменениям в объеме и цене.

Диаграмма, отображающая средний объем торгов биткойном по годам с 2014 по 2023 год. График визуализирует, как изменялся средний объем торгов биткойном на протяжении нескольких лет

```python
df = pd.read_csv("./Bitcoin Price (2014-2023)_exported.csv")  


df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year

average_volume_by_year = df.groupby('Year')['Volume'].mean()
average_volume_by_year = (average_volume_by_year / 1e6).round(2)

print(average_volume_by_year)

plt.figure(figsize=(10, 6))
average_volume_by_year.plot(kind='bar', color='skyblue') 
plt.title('Объем торгов по годам (в миллионах)')  
plt.xlabel('Год')  
plt.ylabel('Объем торгов (млн)') 
plt.xticks(rotation=45) 
plt.grid(axis='y') 
plt.tight_layout() 

for i, v in enumerate(average_volume_by_year):
    plt.text(i, v + 50, f'{v:.2f}', ha='center', va='bottom') 

plt.show()
```

**Применение метода KMeans**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

numeric_df = df_filtered.select_dtypes(include=[np.number])
scaler = StandardScaler()
scaled_data = scaler.fit_transform(numeric_df)
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(scaled_data)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.xlabel('Количество кластеров')
plt.ylabel('WCSS')
plt.show()

kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(scaled_data)

df_filtered['Cluster'] = kmeans.labels_
```

**Визуализация кластеров**

Визуализируем кластеры с помощью PCA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

numeric_df = df_filtered.select_dtypes(include=[np.number])

scaler = StandardScaler()
scaled_data = scaler.fit_transform(numeric_df)

pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_data)

pca_df = pd.DataFrame(data=pca_result, columns=['PCA1', 'PCA2'])
pca_df['Cluster'] = df_filtered['Cluster']

plt.figure(figsize=(10, 6))
n_clusters = df_filtered['Cluster'].nunique()
print(f'Количество уникальных кластеров: {n_clusters}')
for cluster in range(n_clusters):
    plt.scatter(pca_df[pca_df['Cluster'] == cluster]['PCA1'],
                pca_df[pca_df['Cluster'] == cluster]['PCA2'],
                label=f'Cluster {cluster}')

plt.title('Визуализация кластеров с помощью PCA')
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.legend()
plt.grid()
plt.show()
```
